```python
#importar libreria, cargar documento csv desde drive y desplegar las primeras cinco filas del documento
import pandas as pd
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vSWIoDvkRqQIdhSqh7soa6-Nwm1osP5RJlCADfqOQC95qiIvZUYZV1uDvFF9Oo8jg/pub?output=csv"
df= pd.read_csv(url)
df.head()
```

```python
#tipo de datos por columna
df.dtypes
```

```python
#Eliminar fila con valores en blanco
df1=df.drop(df.index[0])
df1.reset_index()
df1.sort_values(by = "# Radicado")
df1.head(3)
```

```python

# Eliminar columnas que no aportan valor#
df1.drop(["Asunto","Revisó","Radicó","Of Radica","Estado","Remitente","Responsable","Pre-Radicado","Tipo Asunto","RegistroPor:","Ver","Unnamed: 19","Info SAE","Unnamed: 21","Estado Términos"], axis=1, inplace=True)
```

```python
df1.head(5)
```

```python
#modificar valores con caracteres distintos que hacen que se presente duplicidad en la columna "Elaboró"

for valor in df1['Elaboró']:
    valor_str = str(valor)
    if '/' in valor_str:
        pos = valor_str.find("/")
        new = valor_str[:pos-1]
        df1['Elaboró'].replace({valor_str: new}, inplace=True)


df1.head()



```
